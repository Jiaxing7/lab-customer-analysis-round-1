```python
#Standardize header names.
file3 = file3.rename(columns={'State':'ST',
                             'Gender':'GENDER'})
file3.columns
```

```python
#Rearrange the columns in the dataframe as needed
column_names = file1.columns
column_names
```

```python
file2 = file2[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type', 'Vehicle Class',
       'Total Claim Amount']]
file2.columns
```

```python
file3 = file3[['Customer', 'ST', 'GENDER', 'Education', 'Customer Lifetime Value',
       'Income', 'Monthly Premium Auto', 'Number of Open Complaints', 'Policy Type', 'Vehicle Class',
       'Total Claim Amount']]
file3.columns
```

```python
# Concatenate the three dataframes
data = pd.concat([file1, file2, file3], axis=0)
data
```

```python
# Perform the data cleaning operations mentioned so far in class
# Delete the column education and the number of open complaints from the dataframe.

data1 = data.drop(['Education','Number of Open Complaints'], axis=1)
data1.head()
```

```python
# Correct the values in the column customer lifetime value. They are given as a percent, so multiply them by 100 and change dtype to numerical type.


data1['Customer Lifetime Value'] = data1['Customer Lifetime Value'].str.replace('%', '')
data1['Customer Lifetime Value'] = data1['Customer Lifetime Value'].astype(float)
data1.head()
```

```python
data1['Customer Lifetime Value'] = data1['Customer Lifetime Value']/100
data1.head()
```

```python
# Check for duplicate rows in the data and remove if any.
print(data1.shape)
```

```python
data2 = data1.copy()
data2 = data2.drop_duplicates()
print(data2.shape)
```

```python
# Filter out the data for customers who have an income of 0 or less.
filtered = data2[data2['Income']>0]
filtered
```
